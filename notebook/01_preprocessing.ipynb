```python
import os
import shutil
from pathlib import Path

def aplanar_carpetas_imagenes(base_dir):
    """
    Saca las imágenes de las subcarpetas daytime/night y las mueve
    directamente a train, val y test. Luego elimina las subcarpetas.
    """
    base_path = Path(base_dir)
    splits = ['train', 'val', 'test']

    for split in splits:
        split_path = base_path / split
        if not split_path.exists():
            # Advertencia agregada para detectar errores de ruta más fácilmente
            print(f"⚠️ Advertencia: No se encontró la ruta {split_path}")
            continue

        print(f"Aplanando carpeta: {split}...")

        # Mover archivos de daytime y night a la raiz del split
        for subfolder in ['daytime', 'night']:
            sub_path = split_path / subfolder
            if sub_path.exists():
                # Mover todas las imágenes .jpg
                for img_file in sub_path.glob('*.jpg'):
                    shutil.move(str(img_file), str(split_path / img_file.name))

                # Eliminar la subcarpeta ahora que está vacía
                try:
                    sub_path.rmdir()
                except OSError:
                    print(f"No se pudo eliminar {sub_path} (revisa si tiene archivos ocultos)")

    print("¡Carpetas aplanadas con éxito! Simetría lista para YOLO.")

# Ejecutar el aplanamiento (ruta relativa desde la carpeta notebooks)
# Se ajustó la ruta para apuntar a 'enhanced' basándonos en tu estructura actual
directorio_imagenes_procesadas = '../data/enhanced'
aplanar_carpetas_imagenes(directorio_imagenes_procesadas)
```
