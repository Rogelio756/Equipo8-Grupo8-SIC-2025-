```python
from ultralytics import YOLO
import glob

# 1. Cargar TU modelo recién entrenado (usamos el archivo best.pt que tiene los mejores pesos)
ruta_pesos = '../models/entrenamiento_script_estable/weights/best.pt'
model = YOLO(ruta_pesos)

# 2. Tomar las primeras 10 imágenes de tu carpeta de pruebas (test)
imagenes_test = glob.glob('../data/enhanced/images/test/*.jpg')[:10]

print("Realizando inferencia en imágenes desconocidas...")

# 3. Hacer la predicción y guardar las imágenes resultantes
resultados = model.predict(
    source=imagenes_test, 
    save=True,                   # Le decimos que dibuje y guarde las imágenes
    project='../reports',        # Guardaremos esto en la carpeta de reportes
    name='predicciones_test',    # Nombre de la subcarpeta
    conf=0.5                     # Solo mostrar predicciones con más del 50% de seguridad
)

print("Imagenes procesadas y guardadas")
```
