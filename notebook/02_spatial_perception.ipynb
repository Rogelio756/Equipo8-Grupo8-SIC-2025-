```python
from ultralytics import YOLO

if __name__ == '__main__':
    # 1. Definir ruta
    ruta_proyecto = ''
    
    # 2. Cargar modelo
    model = YOLO('yolov8n.pt')
    
    print("Iniciando entrenamiento con ajustes de estabilidad para Windows...")
    
    # 3. Entrenar
    results = model.train(
        data=f'{ruta_proyecto}/data.yaml',         
        epochs=100, # Subimos las épocas porque le agregaremos el freno automático
        imgsz=640,                                 
        batch=16,   # Bajamos a 16 para asegurar que no use RAM compartida
        workers=4,  # Bajamos a 4: Evita bloqueos (deadlocks) en Windows
        device=0,
        patience=15, # Early Stopping: Frena si no mejora en 15 épocas
        project=f'{ruta_proyecto}/models',   
        name='entrenamiento_script_estable'         
    )
    
    print("¡Proceso finalizado con éxito!")
```
